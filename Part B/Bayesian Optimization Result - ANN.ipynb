```python
import tensorflow as tf
from functions import train_predict_nn
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.optimizers import Adam
```

```python
batch_size=324.5
learning_rate= 0.0006683
n_epochs=34.05
```

```python
def ann(input_shape):
    ann = Sequential()
    
    units1=467.4
    units2=484.5
    units3=421.6
    units4=496.0
    
    ann = Sequential()
    ann.add(Dense(units=int(units1), activation='relu', input_shape=(324,)))
    ann.add(Dense(units=int(units2), activation='relu'))
    ann.add(Dense(units=int(units3), activation='relu'))
    ann.add(Dense(units=int(units4), activation='relu'))
    ann.add(Dense(units=10, activation='softmax'))
    
    return ann

ann = ann((324, ))
ann.compile(optimizer=Adam(learning_rate=learning_rate), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```
